You are given an array X of floating-point numbers x1, x2, ... xn. These can be rounded up or down to create a corresponding array Y of integers y1, y2, ... yn.

Write an algorithm that finds an appropriate Y array with the following properties:

The rounded sums of both arrays should be equal.

The absolute pairwise difference between elements is minimized. In other words, |x1- y1| + |x2- y2| + ... + |xn- yn| should be as small as possible.
For example, suppose your input is [1.3, 2.3, 4.4]. In this case you cannot do better than [1, 2, 5], which has an absolute difference of |1.3 - 1| + |2.3 - 2| + |4.4 - 5| = 1.

Question from dailycodingproblem.com

```python
import math
```

# Greedy Rounding Algorithm

```python
#Get the sum for rounding all up vs all down
def greedy_rounding(X):
    down_sum = sum(math.floor(x) for x in X)
    up_sum = sum(math.ceil(x) for x in X)
    #Start by finding the sum of the array
    x_sum = round(sum(X))

    #To find the number of values to round up
    up_ct = x_sum - down_sum
    
    #sort the array by magnitude of the remainder between
    #the original number and rounding down
    #to preserve the original array use the sorted function
    sortX = sorted(X, key= lambda x : math.floor(x)-x)
    
    #Create an array to hold the new values
    Y = []
    
    #round up the first N numbers where N = up_ct
    for i in range(up_ct):
        Y.append(math.ceil(sortX[i]))
    
    #round down the remaining values
    for i in range(up_ct, len(sortX)):
        Y.append(math.floor(sortX[i]))
    
    print("Original Array: ", X)
    print("Rounded Array: ",Y)
```

```python
#a set of 10 test cases generated by chat GPT
test_cases = [
    # Test Case 1: Example from the problem statement
    [1.3, 2.3, 4.4],  # Expected Output: [1, 2, 5]

    # Test Case 2: All round down
    [1.1, 2.1, 3.1],  # Expected Output: [1, 2, 3]

    # Test Case 3: All round up
    [1.9, 2.9, 3.9],  # Expected Output: [2, 3, 4]

    # Test Case 4: Mix of rounding up and down to meet sum
    [1.4, 2.6, 3.4],  # Expected Output: [1, 3, 3]

    # Test Case 5: Numbers very close to their integer values
    [2.0001, 3.9999, 5.0001],  # Expected Output: [2, 4, 5]

    # Test Case 6: Large input with varied decimal places
    [1.2, 3.7, 4.8, 2.4, 5.5],  # Expected Output: [1, 4, 5, 2, 5]

    # Test Case 7: Edge case with all equal fractions
    [1.5, 2.5, 3.5, 4.5],  # Expected Output: [1, 2, 4, 5] or [2, 3, 3, 5]

    # Test Case 8: All integers (no rounding needed)
    [1.0, 2.0, 3.0, 4.0],  # Expected Output: [1, 2, 3, 4]

    # Test Case 9: Large numbers with tiny decimal parts
    [1000.01, 2000.02, 3000.03, 4000.04, 5000.05],  # Expected Output: [1000, 2000, 3000, 4000, 5001]

    # Test Case 10: Negative numbers
    [-1.3, -2.6, -3.4],  # Expected Output: [-1, -3, -3]
]
for t in test_cases:
    greedy_rounding(t)
```
